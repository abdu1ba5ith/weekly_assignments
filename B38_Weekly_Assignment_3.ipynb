```python
# Creating Population graph of 5 countries using Matplotlib
```

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
pop = pd.read_csv("countries (2).csv")

pop
```

```python
india = pop[pop.country == 'India']

india

```

```python
plt.plot(india.year,india.population)
plt.xlabel('Year')
plt.ylabel('Population')
plt.show()
```
